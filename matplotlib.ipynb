# Matplotlib
- matplotlib.pyplot is the package used

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,2])
plt.ylabel("Number")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.show()
```

```python
plt.plot([1,2,3,4] , [1,4,9,16])
plt.ylabel('square')
plt.xlabel('numbers')
plt.grid()    # makes grid on
plt.show()
```

```python
plt.plot([1,2,3,4] , [1,4,9,16] , 'ro')
plt.ylabel('square')
plt.xlabel('numbers')
plt.grid()    # makes grid on
plt.show()
```

# Controlng line properties

```python
def add(x , y):
    return (x+y)

a = add(3,10)
print(a)
```

```python
x = [1 , 2 , 3 , 4]
y = [1 , 4 , 9 , 16]
plt.plot(x , y ,linewidth = 5.0)
plt.plot()
```

## using Keyword sept()

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1,y1,x2,y2)

# x1,y1 = lines[0] and x2,y2 = lines[1]

plt.setp(lines[0] , color = 'r' , linewidth = 2.0)

plt.setp(lines[1] , 'color' , 'g' , 'linewidth' , 2.0)

plt.grid()
```

# working with multiple figures and axes

```python
import numpy as np
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0 , 5.0 , 0.1)
t2 = np.arange(0.0 , 5.0 , 0.02)

plt.figure(1)

# subplot(211) 2 represents the row , 1 represent the column and 1 represent the fignum

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1) , 'b--')


plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2) , 'r--')
plt.show()
```
