# Binary Search

```python
import math

def binsea(arr,l,r,x):
    while l <= r:
 
        mid = l + (r - l) // 2
 
       
        if arr[mid] == x:
            return mid
 
       
        elif arr[mid] < x:
            l = mid + 1
 
        
        else:
            r = mid - 1
 
    
    return -1
arr = [1,2,3,4,5,6,7,8]
x = 8
print(binsea(arr,0,len(arr)-1,x))
```

```python
# find elements common in two list

l1 = list(range(100))
random.shuffle(l1)

l2 = list(range(50))
random.shuffle(l2)


# find common elements in list O(n) time and O(m) space if m<n

smallList = {}
for ele in l2:
    smallList[ele] = 1


# now find the common element

cnt = 0
for i in l1:
    if smallList.get(i) != None:
        print(i)
        cnt = cnt+1
        
print("the number od common elements is",cnt)
```
